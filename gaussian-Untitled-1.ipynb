```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
c = 3e8  # Speed of light in vacuum (m/s)
L = 1000  # Length of the optical fiber (m)
beta2 = -2.17e-26  # Chromatic dispersion parameter (s^2/m)
T0 = 1e-12  # Initial pulse width (s)
N = 1024  # Number of points in time domain
Tmax = 50e-12  # Time window (s)

# Time and frequency domains
t = np.linspace(-Tmax, Tmax, N)
dt = t[1] - t[0]
f = np.fft.fftfreq(N, dt)

# Initial Gaussian pulse
A0 = np.exp(-0.5 * (t / T0) ** 2)

# Frequency domain representation of the pulse
A0_f = np.fft.fft(A0)

# Dispersion operator
D_1 = np.exp(-1j * 0.5 * beta2 * (2 * np.pi * f) ** 2 * L)

# Apply dispersion
A1_f = D_1 * A0_f

# Inverse Fourier transform to get the time domain pulse after dispersion
A1 = np.fft.ifft(A1_f)

# Plotting
plt.figure(figsize=(10, 4))
plt.plot(t * 1e12, np.abs(A0) ** 2, label='Input Pulse')
plt.plot(t * 1e12, np.abs(A1) ** 2, label='Output Pulse')
plt.xlabel('Time (ps)')
plt.ylabel('Intensity (a.u.)')
plt.ylim(0,1)
plt.title('Optical Pulse Stretching by Chromatic Dispersion')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 4))
plt.plot(t * 1e12, np.angle(A0) ** 2, label='Input Pulse')
plt.plot(t * 1e12, np.angle(A1) ** 2, label='Output Pulse')
plt.xlabel('Time (ps)')
plt.ylabel('Intensity (a.u.)')
plt.title('Optical Pulse Stretching by Chromatic Dispersion')
plt.legend()
plt.grid()
plt.show()
```

```python
# Dispersion operator
D_2 = np.exp(-1j * 0.5 * (-1.0)*beta2 * (2 * np.pi * f) ** 2 * L)

A2_f = D_2 * A1_f

# Inverse Fourier transform to get the time domain pulse after dispersion
A2 = np.fft.ifft(A2_f)

# Plotting
plt.figure(figsize=(10, 4))
plt.plot(t * 1e12, np.abs(A1) ** 2, label='Input Pulse')
plt.plot(t * 1e12, np.abs(A2) ** 2, label='Output Pulse')
plt.xlabel('Time (ps)')
plt.ylabel('Intensity (a.u.)')
plt.ylim(0,1)
plt.title('Optical Pulse Stretching by Chromatic Dispersion')
plt.legend()
plt.grid()
plt.show()
```
